```python
from sqlalchemy import create_engine
import sys, os
import pandas as pd

constr = 'mysql+pymysql://{user}:{pwd}@{host}'.format(user=os.environ['MYSQL_USERNAME'],
                                                      pwd=os.environ['MYSQL_PASSWORD'],
                                                      host=os.environ['MYSQL_HOST'])
con = create_engine(constr)
```

```python
q_uk_pages = """
SELECT uk_page.page_id as uk_id, 
       uk_page.page_title as uk_title
from ukwiki_p.page as uk_page   
"""
df_uk_pages = pd.read_sql(q_uk_pages, con)
```

```python
print(df_uk_pages.shape)

df_uk_pages.head(10)
#df_uk.to_csv("df_uk.csv")
```

```python
q_uk_last_rev = """
SELECT uk_revs.rev_page as uk_id,
       uk_revs.rev_timestamp as uk_timestamp
       
from ukwiki_p.revision as uk_revs 
where uk_revs.rev_parent_id=0
"""
df_uk_last_rev = pd.read_sql(q_uk_last_rev, con)
```

```python
print(df_uk_last_rev.shape)

df_uk_last_rev.head(10)
#df_en.to_csv("df_en.csv")
```

```python
q_en_pages = """
SELECT en_page.page_id as en_id, 
       en_page.page_title as en_title
from enwiki_p.page as en_page   
"""
df_en_pages = pd.read_sql(q_en_pages, con)
```

```python
print(df_en_pages.shape)

df_en_pages.head(10)
#df_uk.to_csv("df_uk.csv")
```

```python
q_same = """
SELECT db1.ips_site_page as uk_title,
       db2.ips_site_page as en_title
FROM  wikidatawiki_p.wb_items_per_site as db1
INNER JOIN  wikidatawiki_p.wb_items_per_site as db2 
    ON db1.ips_item_id=db2.ips_item_id and 
    db1.ips_site_id = 'ukwiki' and  
    db2.ips_site_id = 'enwiki'
    
"""


df_same = pd.read_sql(q_same, con)
```

```python
print(df_same.shape)

df_same.head(10)
#df_same.to_csv("df_same.csv")
```

```python
print("df_same ", df_same.shape)
print("df_en ", df_en.shape)

df_same_en = pd.merge(df_same, df_en, on="en_title")
print("df_same_en ", df_same_en.shape)
#df_same_en.head(100)

df_same_en_uk = pd.merge(df_same_en, df_uk, on="uk_title")
print("df_same_en_uk ", df_same_en_uk.shape)
```

```python
print(df_same_en_uk.shape)
df_same_en_uk.head(1000)
```

```python
from datetime import datetime
from datetime import timedelta

def get_time(time_str):
    str_time = str( eval(time_str) )
    return datetime.strptime(str_time, '%Y%m%d%H%M%S')
```

```python
for index, row in df_same_en_uk.iterrows():
    en_time = get_time(row['en_timestamp'])
    
    uk_time = get_time(row['uk_timestamp'])
    
    if en_time - uk_time > timedelta(days=30):
        print(row['en_title'])
        print(en_time)
        print(uk_time)
```
