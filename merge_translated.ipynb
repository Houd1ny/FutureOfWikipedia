```python
df = pd.merge(df_translated_uk, df_en, on="en_title")[["uk_id", "uk_title", "uk_timestamp", 
                                                       "en_id", "en_title", "en_timestamp"]]
print(df.shape)

df.head(10)
```

```python
from datetime import datetime
from datetime import timedelta

def get_time(time_str):
    str_time = str( eval(eval(time_str)) )
    return datetime.strptime(str_time, '%Y%m%d%H%M%S')
```

```python
df_clean = pd.DataFrame(columns=["uk_id", "uk_title", "uk_timestamp", 
                                "en_id", "en_title", "en_timestamp"])
for index, row in df.iterrows():
    en_time = get_time(row['en_timestamp'])
    
    uk_time = get_time(row['uk_timestamp'])
    
    if en_time - uk_time > timedelta(days=30):
        df_clean = df_clean.append(row)
        #print(row['en_title'])
        #print(en_time)
        #print(uk_time)

print(df_clean.shape)
df_clean.head(200)
```
